### Observations

##### These results are based on 100,000 simulations of the dartsim adaption manager. 
##### For random simulation of random latencies, the value was limited to 5 periods. The defualt value for latency is 1 period.
##### The most visible differences between the two sets of simulations are the success and destroyed rate. Having random values for latencies negatively affects the performance of the adaption manager.

##### Latency = 1 vs. Latency = (random)
##### Success rate = 0.26827 vs. 0.12586
